QUEUE DENGAN LIST

```python
# Buat variabel buat nampung queue
queue = []

# Tambahkan elemen ke queue
queue.append('a')
queue.append('i')
queue.append('u')
queue.append('e')
queue.append('o')

# Tampilkan queue
print(queue)

# pop() Untuk mengeluarkan data yang di dalam queue
print("\nData yang dikeluarkan: ")
print(queue.pop(0))
print(queue.pop(0))
print(queue.pop(0))

print("\nData yang tersisa di dalam queue: ")
print(queue)
```

QUEUE DENGAN COLLECTIONS .dequeue

```python
# import deque dari collections
from collections import deque

queue = deque()

# masukin data ke queue
queue.append('a')
queue.append('i')
queue.append('u')
queue.append('e')
queue.append('o')

print('isi queue sekarang: ')
print(queue)

# popleft() Untuk mengeluarkan data secara LIFO
print("Data yang dikeluarkan dari queue: ")
print(queue.popleft())
print(queue.popleft())
print(queue.popleft())

# menampilkan data setelah dikeluarkan
print("\nData yang tersisa di dalam queue: ")
print(queue)
```

QUEUE DENGAN queue.queue

```python
#  import Queue daru queue
from queue import Queue

# IInisiasi queue
qu = Queue(maxsize=5)

# qsize()
print(qu.qsize())

# put()
qu.put('a')
qu.put('i')
qu.put('u')
qu.put('e')
qu.put('o')

print("Full: ", qu.full())
print("Size: ", qu.qsize())

# get()
print("\nElements dequeued from the queue")
print(qu.get())
print(qu.get())
print(qu.get())

print("\nEmpty: ", qu.empty())
print("Full: ", qu.full())
```

QUEUE DENGAN SINGLE LINKED LIST

```python
class Queue:
  def __init__(self):
    self.queue = list()

  def addtoqueue(self,dataval):
    # metode untuk memasukan data
    if dataval not in self.queue:
      self.queue.insert(0,dataval)
      return True
    return False

  # metode pop untuk menghapus data
  def removefromq(self):
    if len(self.queue)>0:
      return self.queue.pop()
    return ("Gaada data di dalam queue")

  def size(self):
    return len(self.queue)

TheQueue = Queue()
print(TheQueue.addtoqueue("Jan"))
print(TheQueue.addtoqueue("Feb"))
print(TheQueue.addtoqueue("Mar"))
print(TheQueue.addtoqueue("Apr"))
print(TheQueue.size())
print(TheQueue.removefromq())
print(TheQueue.removefromq())
print(TheQueue.size())
```
