STACK DENGAN LIST

```python
# Buat variabel kosong untuk menampung data dalam stack
stack = []

# append() fungsi untuk memasukan element ke stack
stack.append('i')
stack.append('u')
stack.append('e')
stack.append('o')

print('Initial stack')
print(stack)

# pop() fungsi untuk ngeluarin secara LIFO
print('nElements popped from stack:')
print(stack.pop())
print(stack.pop())
print(stack.pop())

print('\nstack after elements are popped:')
print(stack)
```

STACK DENGAN COLLECTIONS .deque

```python
# Impor library deque dari collections
from collections import deque

# Buat variabel kosong untuk menampung data dalam stack
stack = deque()

# append() ffungsi untuk memasukan data
stack.append('a')
stack.append('i')
stack.append('u')
stack.append('e')
stack.append('o')

print('Initial stack:')
print(stack)

# pop() fungsi untuk mengeluarkan data
print('\nElements popped from stack:')
print(stack.pop())
print(stack.pop())
print(stack.pop())

print("\nstack after elements are popped:")
print(stack)
```

STACK DENGAN queue.LifoQueue

```python
# Impot lifoqueue dari queue
from queue import LifoQueue

# Inisiasi stack
stack = LifoQueue(maxsize=5)

# qsize() digunakan untuk menampilkan angka di dalam element stack
print(stack.qsize())

# put() digunakan untuk menambahkan element ke dalam stack
stack.put('a')
stack.put('i')
stack.put('u')
stack.put('e')
stack.put('o')

print("Full: ", stack.full())
print("Size: ", stack.qsize())

# get() sama seperti pop() digunakan untukj mengambbil data dari stack secara LIFO
print("\nElements yang dikeluarin dari stack: ")
print(stack.get())
print(stack.get())
print(stack.get())

print("\nEmpty: ", stack.empty())
```

STACK DENGAN SINGLE LINKED LIST

```python
# Buat node
class Node:
  def __init__(self, value):
    self.value = value
    self.next = None

class stack:

  # inisiasi stack pakai node dummy
  def __init__(self):
    self.head = Node("Head")
    self.size = 0

  # representasi string pada stack
  def __str__(self):
    cur = self.head.next
    out = ""
    while cur:
      out += str(cur.value) + "->"
      cur = cur.next
    return out[:-2]

  # menambil size stack terkini
  def getSize(self):
    return self.size

  # validasi stack kosong/tidak
  def isEmpty(self):
    return self.size == 0

  # Mengambl stack paling atas
  def peek(self):
    if self.isEmpty():
      raise Exception("Peeking from an empty stack")
    return self.head.next.value

  # Memasukan isi ke stack
  def push(self, value):
    node = Node(value)
    node.next = self.head.next
    self.head.next = node
    self.size += 1

  # Menghapus isi dari stack dan mengembalikannya
  def pop(self):
    if self.isEmpty():
      raise Exception("Popping from an empty stack")
    remove = self.head.next
    self.head.next = self.head.next.next
    self.size -= 1
    return remove.value

# driver code
if __name__ == "__main__":
  stack_obj = stack()
  for i in range(1, 11):
    stack_obj.push(i)
  print(f"stack: {stack_obj}")

  for _ in range(1, 6):
    remove = stack_obj.pop() #
    print(f"Pop: {remove}")
  print(f"stack: {stack_obj}")
```

Implementasi Nyata

```python
print("=" * 70)
print("STUDI KASUS 1: FITUR UNDO/REDO PADA APLIKASI PENGOLAH KATA")
print("=" * 70)

class TextEditor:
    def __init__(self):
        self.undo_stack = []      # Stack untuk menyimpan history
        self.redo_stack = []      # Stack untuk menyimpan yang di-undo
        self.text = ""

    def tulis(self, new_text):
        """Menambahkan teks baru"""
        print(f"\n✍️ Mengetik: '{new_text}'")
        self.undo_stack.append(self.text)  # Simpan state sebelum perubahan
        self.text += new_text
        self.redo_stack.clear()  # Kosongkan redo karena ada perubahan baru
        self.tampilkan_status()

    def undo(self):
        """Membatalkan perubahan terakhir"""
        if not self.undo_stack:
            print("\n❌ Tidak ada yang bisa di-undo!")
            return

        print("\n↩️ Melakukan UNDO")
        self.redo_stack.append(self.text)  # Simpan ke redo untuk kemungkinan redo
        self.text = self.undo_stack.pop()  # Kembalikan ke state sebelumnya
        self.tampilkan_status()

    def redo(self):
        """Mengulangi perubahan yang sudah di-undo"""
        if not self.redo_stack:
            print("\n❌ Tidak ada yang bisa di-redo!")
            return

        print("\n↪️ Melakukan REDO")
        self.undo_stack.append(self.text)  # Simpan ke undo
        self.text = self.redo_stack.pop()  # Ambil dari redo
        self.tampilkan_status()

    def tampilkan_status(self):
        print(f"📄 Teks saat ini: '{self.text}'")
        print(f"📚 Undo stack: {self.undo_stack}")
        print(f"📖 Redo stack: {self.redo_stack}")

# Simulasi penggunaan
editor = TextEditor()
editor.tulis("Halo")
editor.tulis(" dunia")
editor.tulis("!")
editor.undo()
editor.undo()
editor.redo()
editor.tulis(" Apa kabar?")
```
